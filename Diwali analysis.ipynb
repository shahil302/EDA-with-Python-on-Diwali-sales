```python
# import python libraries

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode escape')
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'], axis =1 , inplace = True )
```

```python
 df.isnull()
```

```python
# checking null values
df.isnull().sum()
```

```python
# Droping null values
df.dropna(inplace = True)
```

```python
# change datatypes
df['Amount'] = df['Amount'].astype('int')
```

```python
# here we don't use inplace = true so that it doesn't save as we just use this to learn 
df.rename(columns={'Marital_Status':'saadi'})
```

# Exploratory Data Analysis

### Gender


```python
ax = sns.countplot(x= 'Gender' , data =df)
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = df.groupby('Gender', as_index = False)['Amount'].sum()
sns.barplot(x = 'Gender' , y = 'Amount' , data = ax)
```

#### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men 

```python
ax = sns.countplot(x= 'Age Group', data = df, hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax = df.groupby('Age Group', as_index = False)['Amount'].sum()
sns.barplot(x='Age Group', y= 'Amount', data = ax)
```

```python
ax = df.groupby('Age Group', as_index = False)['Amount'].sum().sort_values(by='Amount', ascending= False)
sns.barplot(x='Age Group', y= 'Amount', data = ax)
```

###### From above graphs we can see that most of the buyers are of age group between 26-35 yrs female


### States

```python
ax = df.groupby('State', as_index= False)['Orders'].sum().sort_values(by='Orders',ascending= False)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='State', y='Orders', data= ax)
```

```python
ax = df.groupby('State', as_index= False)['Orders'].sum().sort_values(by='Orders',ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Orders', data= ax)
```

```python
ax = df.groupby('State', as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State', y='Amount', data= ax)
```

#### From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital status

```python
ax = sns.countplot(x= 'Marital_Status' , data = df)

sns.set(rc={'figure.figsize':(6,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status', y='Amount', data= ax, hue = 'Gender')
```

#### From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

## Occupation

```python
ax = sns.countplot(x= 'Occupation' , data = df)

sns.set(rc={'figure.figsize':(26,7)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax = df.groupby('Occupation', as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False)

sns.set(rc={'figure.figsize':(26,7)})
sns.barplot(x='Occupation', y='Amount', data= ax)
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

### Product category

```python
ax = sns.countplot(x= 'Product_Category' , data = df)

sns.set(rc={'figure.figsize':(26,7)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
ax = df.groupby('Product_Category', as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='Product_Category', y='Amount', data= ax)
```

```python
ax = df.groupby('Product_ID', as_index= False)['Orders'].sum().sort_values(by='Orders',ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='Product_ID', y='Orders', data= ax)
```
